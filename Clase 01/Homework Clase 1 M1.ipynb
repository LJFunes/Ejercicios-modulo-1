1) Crear una función capaz de convertir números enteros de base 10 a base 2. Debe recibir como parámetro el número a convertir<br>
Consideraciones:<br> 
a. Tratar de resolverlo sin usar la función format(nro,"b")<br>

```python
def NumeroBinario(numero):
    '''
    Esta función recibe como parámetro un número entero mayor ó igual a cero y lo devuelve en su 
    representación binaria. Debe recibir y devolver un valor de tipo entero.
    En caso de que el parámetro no sea de tipo entero y mayor a -1 retorna nulo.
    '''
    if type(numero)==int and numero>=0:
        strbin = ""
        while numero > 0:
            strbin = str(numero%2) + strbin
            numero = numero//2
        if strbin == "": return 0
        return int(strbin)
    else: return None
```

```python
#Ejemplo: 11101
NumeroBinario(29)
```

```python
def funcionBinaria1(num):
    if type(num)==int:
        listabin=list(bin(num))
        listabin.remove("b")
        return int("".join(listabin))
    else: return "El valor ingresado debe ser un numero entero"
```

```python
funcionBinaria1(-10)
```

```python
def funcionBinaria2(num):
    if type(num)==int:
        return int(format(num,"b"))
    else: return "El valor ingresado debe ser un numero entero"
```

Salida esperada:

* Fraccion 1 / 8 :  0.125  y En binario:  0.001

```python
def fraccionBinaria(num):
    if type(num)==float and num>=0 and num<=1:
        strbin = ""
        for rep in range(24):
            strbin = strbin + str(int(num*2))
            num = num*2 - int(num*2)
            if num == 0:
                break
        return "0."+strbin
    else: None
```
