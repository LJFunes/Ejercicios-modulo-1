1) Abrir el archivo "Emisiones_CO2.csv" y cargar sus datos en un diccionario.

```python
dicc_emisiones = {  'cod_pais' : [],
                    'nom_pais' : [],
                    'region' : [],
                    'anio' : [],
                    'co2' : [],
                    'co2_percapita' : []}
```

```python
import os
archivo = open('Emisiones_CO2.csv') # viene por defecto "r" y decoding=ANSI o latin-8)
next(archivo)
for linea in archivo:
    linea = linea.strip()
    lista = linea.split("|")
    dicc_emisiones["cod_pais"].append(lista[0])
    dicc_emisiones["nom_pais"].append(lista[1])
    dicc_emisiones["region"].append(lista[2])
    dicc_emisiones["anio"].append(lista[3])
    dicc_emisiones["co2"].append(lista[4])
    dicc_emisiones["co2_percapita"].append(lista[5])
archivo.close()
```

```python
if "" in dicc_emisiones["cod_pais"]:
    print("hay faltantes en cod_pais")
if "" in dicc_emisiones["nom_pais"]:
    print("hay faltantes en nom_pais")
if "" in dicc_emisiones["region"]:
    print("hay faltantes en region")
if "" in dicc_emisiones["anio"]:
    print("hay faltantes en anio")
if "" in dicc_emisiones["co2"]:
    print("hay faltantes en co2")
if "" in dicc_emisiones["co2_percapita"]:
    print("hay faltantes en co2_percapita")
```

```python
for ind, anio in enumerate(dicc_emisiones["anio"]):
    anio = int(anio)
    dicc_emisiones["anio"][ind]=anio
```

```python
# ¿Cuál es el total de emisiones de CO2 para 'América Latina y Caribe' en el año 2010?

for ind, co2 in enumerate(dicc_emisiones["co2"]):
    if co2 != "":
        co2 = co2.replace(".","").replace(",",".")
        co2 = float(co2)
    else:
        co2 = None
    dicc_emisiones["co2"][ind]=co2   
```

```python
for ind, co2 in enumerate(dicc_emisiones["co2_percapita"]):
    if co2 != "":
        co2 = co2.replace(".","").replace(",",".")
        co2 = float(co2)
        co2 = round(co2,2)
    else:
        co2 = None
    dicc_emisiones["co2_percapita"][ind]=co2
```

```python
emisiones=0
for ind, claves in enumerate(dicc_emisiones["region"]):
    if (dicc_emisiones["region"][ind] == "América Latina y Caribe"):
        if (dicc_emisiones["anio"][ind] == 2010):
            if (dicc_emisiones["co2"][ind] != None):
                emisiones += dicc_emisiones["co2"][ind]

emisiones = round(emisiones,2)

print(f'El total de emisiones de CO2 para América Latina y Caribe en el año 2010 fue de {emisiones} kt')
```
